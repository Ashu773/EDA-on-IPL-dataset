```python
import matplotlib.pyplot as plt

winner_wins = df.groupby('winner')['win_by_runs'].sum()

plt.bar(winner_wins.index, winner_wins.values)
plt.xlabel('Winner')
plt.ylabel('Total Win by Runs')
plt.title('Total Win by Runs for Each Team')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python


df_toss_winner_winner = df[['toss_winner', 'winner']]

grouped_df = df_toss_winner_winner.groupby(['toss_winner', 'winner']).size().unstack()

fig, ax = plt.subplots(figsize=(10, 7))

grouped_df.plot(kind='bar', stacked=True, ax=ax)

ax.set_xlabel('Toss Winner')
ax.set_ylabel('Number of Matches Won')
ax.set_title('Toss Winner vs Match Winner')



plt.show()
```
